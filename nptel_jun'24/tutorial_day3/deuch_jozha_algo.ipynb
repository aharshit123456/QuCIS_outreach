In 1985, a British physicist at the University of Oxford devised the first quantum computational algorithm; one that tackles a seemingly unimportant/contrived problem. Although it has no use in today’s realm of QC, this primal algorithm, nonetheless, serves as an excellent proof of concept that, in certain instances, quantum computers can exponentially outperform their classical analogues.

### Deutsch’s Problem

Deutsch-Josza algorithm is used to determine if a function $f: \{0, 1\}^n \to \{0, 1\}$ is constant or balanced, it is to say, we want to determine if $f(x) = 0$ or $f(x) = 1$ for all $x$; or half of our domain gives as result $0$ and the other half give us $1$.

By utilizing the powers of superposition, David Deutsch proved that his problem could be solved in one go — a single quantum oracle that performs a single logical query. Furthermore, the quantum computer can evaluate multiple values of $‘x’$ for $f(x)$ instantaneously.

**Lets get dive inot the solution step by step**

* Consider two input qubits, one initialized in the state, $|0\rangle$, the other in the state, $|1\rangle$. Visualizing this in vector form, the qubits begin as:

### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_0\rangle = |01\rangle$

* From this, we apply a Hadamard transformation to each qubit. This quantum operator performs a one-qubit rotation, converting the basis vectors into superposition states with equal probabilities of measuring zero or one. <br>

* The normalized unitary matrix for the Hadamard transformation is defined below:

#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ H = \dfrac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}$

Thus, our qubits will now be in the state:

#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{1}\rangle = \left( \dfrac{|0\rangle + |1\rangle}{\sqrt{2}} \right) \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}}\right)$

and the expanded form can be return as :
#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{1}\rangle = \dfrac{|00\rangle - |01\rangle + |10\rangle - |11\rangle}{2}$

Now, as the qubits pass through the unitary matrix that maps $|x, y\rangle \rightarrow |x, y ⊕ f(x)\rangle$ , they are acted on as follows:

#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \dfrac{|0, 0 \oplus f(0)\rangle - |0, 1 \oplus f(0)\rangle + |1, 0 \oplus f(1)\rangle - |1, 1 \oplus f(1)\rangle}{2}$

Let’s expand upon this oracle in a case-by-case basis, in order to gain better intuition of what is truly going on in this algorithm.

* ### Let f(0) = 0, f(1) = 0.

#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \dfrac{|00\rangle - |01\rangle + |10\rangle - |11\rangle}{2}$
#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \left( \dfrac{|0\rangle + |1\rangle}{\sqrt{2}} \right) \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}}\right)$

* ### Let f(0) = 1, f(1) = 1.





#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \dfrac{|01\rangle - |00\rangle + |11\rangle - |10\rangle}{2}$
#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \left( \dfrac{|0\rangle + |1\rangle}{\sqrt{2}} \right) \left( \dfrac{|1\rangle - |0\rangle}{\sqrt{2}}\right)$
#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = - \left( \dfrac{|0\rangle + |1\rangle}{\sqrt{2}} \right) \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}}\right)$

* ### Let f(0) = 0, f(1) = 1.

#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \dfrac{|00\rangle - |01\rangle + |11\rangle - |10\rangle}{2}$
#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}} \right) \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}}\right)$

* ### Let f(0) = 1, f(1) = 0.


#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \dfrac{|01\rangle - |00\rangle + |10\rangle - |11\rangle}{2}$
#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}} \right) \left( \dfrac{|1\rangle - |0\rangle}{\sqrt{2}}\right)$
#### $\ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ \ \ \  \ \ \ \ \ \  \ |\psi_{2}\rangle = - \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}} \right) \left( \dfrac{|0\rangle - |1\rangle}{\sqrt{2}}\right)$

With the factored results of all four possible cases, we must now pass the first qubit through a Hadamard gate <br>once again and measure the values. This final step of the algorithm yields the following:

$$
|\psi_2\rangle = \left\{\begin{array}{ll}
  \pm |0\rangle \left(\dfrac{|0\rangle - |1\rangle}{\sqrt{2}}\right) & \text{if  } f(0) = f(1) \\
 \pm|1\rangle \left(\dfrac{|0\rangle - |1\rangle}{\sqrt{2}}\right) & \text{if  } f(0) \neq f(1) \\
\end{array}\right.
$$

As you can see, for cases one and two — where $f(0) = f(1)$ — Deutsch’s algorithm does not change the state of the first qubit. Whereas for cases three and four — where $f(0) \neq f(1)$ — the first qubit is flipped from $|0\rangle \rightarrow |1\rangle$. Contrary to classical means of solving Deutsch’s problem, we can determine a character of two functions be only performing one evaluation.

### Deutsch’s Algorithm in PennyLane

```python
## Install the required packages with the following commands(do uncommnet the commands)

# pip install pennylane
# pip install numpy

## import the circuit

import pennylane as qml
import numpy as np
```

```python
## create a device
dev = qml.device("default.qubit", wires=2, shots=1)
```

Now, we need to define the function that will be evaluated (one of the four cases explained in the previous section). <br>Let's consider the first case, where $f(0) = 0 $ and $ f(1) = 0 $.

```python
def f(x):
    #f :{0, 1} --> {0, 1}     
    if x == 0:
        return 0
    if x == 1:
        return 0     
```

```python
## Prepare the oracle

def oracle(g):
    oracle = np.array([
        [1-g(0), g(0), 0, 0],
        [g(0), 1-g(0), 0, 0],
        [0, 0,1-g(1), g(1)],
        [0 ,0 ,g(1),1-g(1)]
    ])
    return oracle
```

Further, let’s define our unitary matrix as the oracle calling the function we defined above.

```python
## Mapping oracle

oracle_f = oracle(f)
```

```python
# Create a Qnode 
@qml.qnode(dev)

def circuit1():
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.QubitUnitary(oracle_f, wires=[0, 1])
    qml.Hadamard(wires=0)
    return qml.sample(wires=0)
```

```python
qml.draw_mpl(circuit)()
```
